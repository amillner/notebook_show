```python
def fastrandsample(p,n=1):
    
    # Multinomial random numbers.
    #
    # USAGE: y = fastrandsample(p,n)
    
    from scipy import stats
    from scipy.stats import binned_statistic
    import numpy as np
    from numpy import random, arange, cumsum, insert
    from numpy.random import rand
    
    bins_ = insert(cumsum(p),0,0)

    _,_,y = binned_statistic(rand(n),arange(n),statistic='count',bins=bins_)

    #because python starts at zero and this will be used for indexing later
    y = y-1

    return(y)
```

```python
def mfit_optimize_no_priors(likfun,param,data,nstarts = 5):
    
    # Find maximum a posteriori parameter estimates.
    #
    # USAGE: results = mfit_optimize(likfun,param,data,[nstarts])
    #
    # INPUTS:
    #   likfun - likelihood function handle
    #   param - [K x 1] parameter structure
    #   data - [S x 1] data structure
    #   nstarts (optional) - number of random starts (default: 5)
    #
    # OUTPUTS:
    #   results - structure with the following fields:
    #               .x - [S x K] parameter estimates
    #               .logpost - [S x 1] log posterior
    #               .bic - [S x 1] Bayesian information criterion
    #               .aic - [S x 1] Akaike information criterion
    #               .H - [S x 1] cell array of Hessian matrices
    #
    # Sam Gershman, June 2015
    from scipy.optimize import minimize
    from numpy import log, zeros
    from numpy.random import uniform
    #from RLfun_modV2 import rllik,rlRW,rlRW_lapse_rate,rlRW_lapse_rate_q0,rlRW_lapse_rate_bias,rlRW_lapse_rate_bias_rew_pun,rlRW_lapse_bias_Pav
    
    S = len(data)
    K = len(param)
    results = {'K':K}

    results['param'] = param
    results['likfun'] = likfun

    bounds = []
    for i in range(K):
        bounds.append((param[i]['lb'],param[i]['ub']))
    bounds = tuple(bounds)
    
    results['logpost'] = zeros((S,nstarts))
    results['x'] = zeros((S,K))
    #results['H'] = zeros((S,K*K))
    results['bic'] = zeros((S,nstarts))
    results['aic'] = zeros((S,nstarts))
    results['winner'] = zeros(S)

    for s in range(S):
        print('Subject %d'%s)
        
        # construct posterior function
        f = lambda x: -likfun(x,data[s])
        for i in range(nstarts):
            x0 = zeros(K)
            for k in range(K):
                x0[k]= uniform(param[k]['lb'],param[k]['ub'])
                if i==2:
                    x0=array([4,.1,.001,5,.01])
            opt_min_result = minimize(f,x0,hess=None,bounds=bounds)
            x = opt_min_result.x
            logp = -opt_min_result.fun
    
            results['logpost'][s][i] = logp
            if i == 0 or results['logpost'][s][i-1] < logp or results['logpost'][s][i-2] < logp:
                
                results['x'][s] = x
                results['winner'][s]=i

            results['bic'][s][i] = K*log(data[s]['N']) - 2*results['logpost'][s][i]
            results['aic'][s][i] = K*2 - 2*results['logpost'][s][i]

    return(results)
```

```python
def sim_hierarchicalEscapeTD(x,trllist,trialns):
    from numpy import linspace,unique,zeros,log,sum,exp,mean,isnan,dot,round
    from numpy.random import choice
    from scipy.stats import beta,expon,gamma
    from scipy.special import beta as _beta
    from scipy.misc import logsumexp
    
    tc = 0 #trial counter (allows skipping to next block)
    block_count=0
    blockList = range(0,1800,30)
    
    
    tmp = x[0]    #temperature for softmax
    rho = x[1]    #coupling parameter
    b0 = x[2]     #top-level bias
    rho0 = x[3]    #top-level coupling
    sw_init = x[4] #initial ratio of values for 3 choices - makes value of a switch low 
    
    n = 20                #grid of bias values
    b = linspace(0.01,0.99,n)  #inductive bias
    
    u = [0,1,2] #unique options (removed unique(cue))
    K = 3 #number of arms (removed len(u))
    A = _beta(rho*b,rho*(1-b)) #Denominator for posterior 
    
    cue=[0,1,2]
       
    #little b0=mean of beta distr which b (bias) is drawn from
    #rho0 = top-level coupling
    B0 = beta.pdf(b,rho0*b0,rho0*(1-b0))
    
    #priors
    score = log(gamma.pdf(tmp,1,5)) + log(gamma.pdf(rho,3,2)) + log(gamma.pdf(rho0,20,1))
    lik = 0
    
    #container for Value after bias included
    V=zeros(K,dtype='f')

    #while tc <= 1200:#use while loop to allow for skipping to next "game"
    while tc < trialns:#use while loop to allow for skipping to next "game"
        gt = trllist.loc[tc,'gt'] #Game trial (trial number within each game)
        if gt == 1: #Reset choice and reward counter when in a new game
            C = zeros((1,K))[0] #Number of times an arm is chosen
            R = zeros((1,K))[0] #Keep track of reward for each arm
            if tc == 0: #if game trial == 1 AND it's not the first trial of the entire task, reset value of option to value of a switch
                v = zeros((n,K)) 
                for j in range(n):
                    v[j,:] = b[j]
        if gt==30:
            block_count+=1

        B = B0;
        for j in range(K):
            B = B * _beta(R[j]+rho*b,C[j]-R[j]+rho*(1-b))/A #posterior over reward possibilities

        w = B/sum(B) # determine weight for value
        
        if tc==0 or (V==0).all():
            V[0]=1
            V[1]=1
            V[2]=1*sw_init
        else:
            V[c] = dot(w,v[:,c])
        
        #collect prob for each option
        trllist.loc[tc,'p1'],trllist.loc[tc,'p2'],trllist.loc[tc,'psw'] = exp(tmp*V - logsumexp(tmp * V,axis=0))
        
        #specify prob for each option in p
        p = exp(tmp*V - logsumexp(tmp * V,axis=0))
        
        if p[0]<.0000001:
            p[0]=.0000001
        if p[1]<.0000001:
            p[1]=.0000001
        if p[2]<.0000001:
            p[2]=.0000001
        
        #make and record choice
        c=fastrandsample(p)[0]
        trllist.loc[tc,'choice']=c
        
        if c == 2:#If choice == 2 (option 3, then skip to next game)
            block_count+=1
            tc=blockList[block_count]
        else:
            r = trllist.iloc[tc,c] #get reward for this choice
            trllist.loc[tc,'rr']=r #record reward

            pe = r - v[:,c]
            lr = 1 / (C[c] + rho + 1)
            C[c] = C[c] + 1
            R[c] = R[c] + r
            v[:,c] = v[:,c] + lr*pe

            tc+=1
        
    return(trllist)
```

```python
def hierarchicalEscapeTD(x,data):
    from numpy import linspace,unique,zeros,log,sum,exp,mean,isnan,dot,round
    from numpy.random import choice
    from scipy.stats import beta,expon,gamma
    from scipy.special import beta as _beta
    from scipy.misc import logsumexp

    c=data['c'].astype(int)
    r=data['r'].astype(int)
    gt=data['gmtn'].astype(int)
    

    tmp = x[0]   #observation noise
    rho = x[1]     #coupling parameter
    b0 = x[2]     #top-level bias
    rho0 = x[3]   #top-level coupling
    sw_init = x[4] #initial ratio of values for 3 choices - makes value of a switch low 
    
    n = 20                #grid of bias values
    b = linspace(0.01,0.99,n)  #inductive bias
    
    u = [0,1,2] #unique options (removed unique(cue))
    K = 3 #number of arms (removed len(u))
    A = _beta(rho*b,rho*(1-b)) #Denominator for posterior 
    
    #b0=mean of beta distr which b (bias) is drawn from
    #rho0 = top-level coupling
    B0 = beta.pdf(b,rho0*b0,rho0*(1-b0))

    #cue=[0,1,2]
    vp = zeros((len(r),3))
    PE = zeros(len(r))
    
    
    #priors
    score = log(gamma.pdf(tmp,1,5)) + log(gamma.pdf(rho,3,2)) + log(gamma.pdf(rho0,20,1))
    lik = 0
    #V=zeros((K,len(r)))
    V=zeros(K,dtype='f')
    
    for i in range(len(r)):
        if ~isnan(c[i]):
            if gt[i] == 1:
                C = zeros((1,K))[0] #Number of times an arm is chosen
                R = zeros((1,K))[0] #Keep track of reward for each arm
                if i == 0: #set v if first trial of whole task 
                    v = zeros((n,K)) #Use PEs to update v
                    for j in range(n):
                        v[j,:] = b[j]

            B = B0
            for j in range(K):
                B = B * _beta(R[j]+rho*b,C[j]-R[j]+rho*(1-b))/A #posterior over reward possibilities
            
            w = B/sum(B)
            if i==0 or (V==0).all():
                V[0]=1
                V[1]=1
                V[2]=1*sw_init
            elif gt[i]==1:
                V[0]=V[0]
                V[1]=V[1]
                V[2]=V[2]*sw_init
            else:
                V[c[i]] = dot(w,v[:,c[i]])

            lik = lik + tmp*V[c[i]] - logsumexp(tmp * V,axis=0)
            
            vp[i] = V

            if c[i]==2:
                continue
            else:
                pe = r[i] - v[:,c[i]]
                lr = 1 / (C[c[i]] + rho + 1)
                C[c[i]] = C[c[i]] + 1
                R[c[i]] = R[c[i]] + r[i]
                v[:,c[i]] = v[:,c[i]] + lr*pe
                
                PE[i] = mean(pe)

    score = score + lik
    #return(score,vp,lik,PE)
    return(lik)
```

```python
def fixtrllist(x):
    fixedtrllist=pd.Series(index=['R1','R2','cond','GameNum','gt','choice','p1','p2','psw','rr'])
    if x.ProbL==90 or x.ProbL==60:
        fixedtrllist.R1=x.RespResultL
        fixedtrllist.R2=x.RespResultR
    elif  x.ProbR==90 or x.ProbR==60:
        fixedtrllist.R1=x.RespResultR
        fixedtrllist.R2=x.RespResultL
    fixedtrllist.cond=x.Cond
    fixedtrllist.GameNum=x.GameNum
    return(fixedtrllist)
```

```python
def hierarchicalEscapeTD(x,r,c):
    from numpy import linspace,unique,zeros,log,sum,exp,mean,isnan,dot
    from scipy.stats import beta,expon,gamma
    from scipy.special import beta as _beta
    from scipy.misc import logsumexp
    
    tmp = x[0]    #temperature for softmax
    rho = x[1]    #coupling parameter
    b0 = x[2]     #top-level bias
    rho0 = x[3]    #top-level coupling
    
    #Need grid to try out different 
    
    n = 20                #grid of bias values
    b = linspace(0.01,0.99,n)  #inductive bias
    
    u = [0,1,2] #unique(cue)
    K = 3 #len(u)
    A = _beta(rho*b,rho*(1-b)) #Arm-#Denominator for posterior 
    
    cue=[0,1,2]
    
    p_c0=zeros((n,len(r)))
    p_c1=zeros((n,len(r)))
    p_sw=zeros((n,len(r)))
    
    #b0=mean of beta distr which (1)  where (1) b (bias) is drawn from and (2) 
    #rho = top-level coupling
    B0 = beta.pdf(b,rho0*b0,rho0*(1-b0))
   
    vp = zeros((len(r)))
    PE = zeros((len(r)))
    
    
    #priors
    score = log(gamma.pdf(tmp,1,5)) + log(gamma.pdf(rho,3,2)) + log(gamma.pdf(rho0,20,1))
    lik = 0
    #V=zeros((K,len(r)))
    V=zeros(K,dtype='f')
    
    for i in range(len(r)):
        if ~isnan(c[i]):
            if gt[i] == 1:
                C = zeros((1,K))[0] #Number of times an arm is chosen
                R = zeros((1,K))[0] #Keep track of reward for each arm
                if i > 0:
                    v[:,0] = v[:,2]
                    v[:,1] = v[:,2]
                else:
                    v = zeros((n,K)) #Use PEs to update v
                    for j in range(n):
                        v[j,:] = b[j]

            B = B0;
            for j in range(K):
                B = B * _beta(R[j]+rho*b,C[j]-R[j]+rho*(1-b))/A #posterior over reward possibilities

            w = B/sum(B)
            

            V[c[i]] = dot(w,v[:,c[i]])

            p_c0[:,i] = exp(tmp*V[0] - logsumexp(tmp * V,axis=0))
            p_c1[:,i] = exp(tmp*V[1] - logsumexp(tmp * V,axis=0))      
            p_sw[:,i]= exp(tmp*V[2] - logsumexp(tmp * V,axis=0))

            lik = lik + tmp*V[c] - logsumexp(tmp * V,axis=0)

            pe = r[i] - v[:,c[i]]
            lr = 1 / (C[c[i]] + rho + 1)
            C[c[i]] = C[c[i]] + 1
            R[c[i]] = R[c[i]] + r[i]
            v[:,c[i]] = v[:,c[i]] + lr*pe
            #Update state/switch
            v[:,2] = v[:,2] + lr*(r[i] - v[:,2])
            
            PE[i] = mean(pe)

    score = score + lik
    return(score,vp,lik,PE)
```
