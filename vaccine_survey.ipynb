```python
d_vac1=d[d['vac1']==1]
d_vac2=d[d['vac2']==1]
```

```python
def gpby_tabl_cnt_pct(df,grpbycol,outcol,reg=True):
    if reg==True:
        t=pd.concat([df.groupby(grpbycol)[outcol].count(),df.groupby(grpbycol)[outcol].count()/df.groupby(grpbycol)[outcol].count().sum()],axis=1)
    elif reg==False:
        t=pd.concat([df.groupby(grpbycol)[outcol].count(),df.groupby(grpbycol)[outcol].count()/df.groupby(grpbycol)[outcol].count().sum(),df.groupby(grpbycol)[outcol].count()/df.groupby(grpbycol[0])[outcol].count()],axis=1)
    return(t)
```
